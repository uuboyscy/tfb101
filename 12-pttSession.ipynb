```python
import requests
from bs4 import BeautifulSoup

headers = {
    'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 11_2_2) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.82 Safari/537.36'
}

landingPageUrl = 'https://www.ptt.cc/ask/over18?from=%2Fbbs%2FGossiping%2Findex.html'
over18Url = 'https://www.ptt.cc/ask/over18'
url = 'https://www.ptt.cc/bbs/Gossiping/index.html'

ss = requests.session()

data = dict()
res = ss.get(landingPageUrl, headers=headers)
soup = BeautifulSoup(res.text, 'html.parser')

'''
[<input name="from" type="hidden" value="/bbs/Gossiping/index.html"/>]
'''
data[soup.select('input[type="hidden"]')[0]['name']] = soup.select('input[type="hidden"]')[0]['value']

'''
[<button class="btn-big" name="yes" type="submit" value="yes">我同意，我已年滿十八歲<br/><small>進入</small></button>,
 <button class="btn-big" name="no" type="submit" value="no">未滿十八歲或不同意本條款<br/><small>離開</small></button>]
'''
data[soup.select('button')[0]['name']] = soup.select('button')[0]['value']

# # Get cookies over18=1
ss.post(over18Url, headers=headers, data=data)

res = ss.get(url, headers=headers)

print(res.text)
```
